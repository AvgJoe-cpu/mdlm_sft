```python
import wandb

# ── Final training configuration (from Stage-B HPO) ──────────────────────
# LR pinned to Stage-B rank-1; other hyperparameters pinned per top-15
# analysis. Configuration is identical for M1 and M2; only substrate and
# step budget differ, and step budgets are compute-matched.
FINAL_CONFIG = {
    # HPO winners
    "learning_rate":   1.84e-3,     # Stage-B rank-1
    "warmup_ratio":    0.1,         # Stage-A top-15 empirical
    "weight_decay":    0.01,        # Stage-A rank-1
    "time_epsilon":    1e-3,        # ε-sweep midpoint

    # Training procedure pins (identical to sweeps)
    "lr_scheduler_type":           "cosine",
    "per_device_train_batch_size": 64,
    "per_device_eval_batch_size":  64,
    "gradient_accumulation_steps": 5,           # B_eff = 320
    "loss_weight_type":            "scheduler",

    # Reproducibility
    "seed":       42,
}

# ── Per-model substrate and step budget ──────────────────────────────────
# Compute-matched: both models train for 4,688 optimizer steps.
# M1: 2 epochs on 50% split (2 × 2344 = 4688)
# M2: 1 epoch on 100% split (1 × 4688 = 4688)
MODEL_SPECS = {
    "M1": {
        "train_ds_path": "/content/cot-strat/strat_train_50pct",
        "max_steps":     4688,
        "eval_steps":    469,        # ~10 evals over 2 epochs
        "hub_repo":      "avgJo3/mdlm-sft-M1-50pct-2ep",
    },
    "M2": {
        "train_ds_path": "/content/cot-strat/strat_train_100pct",
        "max_steps":     4688,
        "eval_steps":    469,        # ~10 evals over 1 epoch
        "hub_repo":      "avgJo3/mdlm-sft-M2-100pct-1ep",
    },
}

FIXED = {
    "load_model_train_path": "/content/mdlm_sft/artifacts_weights_mdlm_base_mdlm-owt_chat",
    "load_data_eval_path":   "/content/cot-strat/validation",
}

# ── Runner ────────────────────────────────────────────────────────────────
def train_and_push(model_tag: str, project: str = "huggingface"):
    """Train M1 or M2 with the final config; push resulting weights to HF Hub."""
    assert model_tag in MODEL_SPECS, f"unknown model_tag: {model_tag}"
    spec = MODEL_SPECS[model_tag]

    run_name = f"mdlm-sft-final-{model_tag}"
    output_dir = f"/model_out_{run_name}"

    print(f"[{run_name}] Starting final SFT run")
    print(f"  substrate:  {spec['train_ds_path']}")
    print(f"  max_steps:  {spec['max_steps']}")
    print(f"  hub_repo:   {spec['hub_repo']}")
    print(f"  config:     {FINAL_CONFIG}")

    with wandb.init(project=project, name=run_name, config={**FINAL_CONFIG, **spec}):
        cfg = MDLMSFTConfig(
            model_name_or_path=FIXED["load_model_train_path"],
            output_dir=output_dir,
            train_ds_path=spec["train_ds_path"],
            eval_ds_path=FIXED["load_data_eval_path"],
            run_name=run_name,
            max_steps=spec["max_steps"],
            eval_steps=spec["eval_steps"],
            # Save final checkpoint locally so we can push it.
            save_strategy="steps",
            save_steps=spec["max_steps"],   # save exactly once at the end
            save_total_limit=1,
            # Push to Hub after training completes.
            push_to_hub=True,
            hub_model_id=spec["hub_repo"],
            hub_strategy="end",             # push once at end, not every save
            hub_private_repo=True,          # start private; make public manually if desired
            **FINAL_CONFIG,
        )
        run_training(cfg)

    print(f"[{run_name}] Complete. Weights pushed to {spec['hub_repo']}.")


# ── Execute one model at a time ───────────────────────────────────────────
# Run these in separate cells so you can inspect each run's outcome before
# starting the next.
#
# Cell 1:  train_and_push("M1")
# Cell 2:  train_and_push("M2")
```

```python
import json
import statistics
from pathlib import Path
import numpy as np

FINAL_EVAL_CONFIG = {
    "max_length":                 1024,
    "per_device_eval_batch_size": 64,
    "dataloader_num_workers":     8,
    "bf16":                       True,
    "num_mc_samples":             1,        # K comes from outer loop
    "time_epsilon":               1e-3,
}

MODEL_CHECKPOINTS = {
    "M0": "/content/mdlm_sft/artifacts_weights_mdlm_base_mdlm-owt_chat",  # or wherever the untrained ref lives
    "M1": "avgJo3/mdlm-sft-M1-50pct-2ep",   # = M+
    "M2": "avgJo3/mdlm-sft-M2-100pct-1ep",  # = M++
}
NUM_SEEDS = 32   # bump from 2 → 16 for the paper numbers

EVAL_DATASETS     = {"test": "/content/cot-strat/test"}
SAVE_ROOT         = "/content/eval_outputs"

_METRIC_KEYS = ("nll", "mean_token_accuracy")


def _mean_se(vs: list[float]) -> dict:
    k = len(vs)
    return {
        "mean":   statistics.mean(vs) if k else None,
        "se":     statistics.stdev(vs) / (k ** 0.5) if k > 1 else 0.0,
        "k":      k,
        "values": vs,
    }


def evaluate_checkpoint(model_tag: str, dataset_tag: str, num_seeds: int = NUM_SEEDS):
    ckpt      = MODEL_CHECKPOINTS[model_tag]
    gold      = EVAL_DATASETS[dataset_tag]
    pair_root = Path(SAVE_ROOT) / f"{model_tag}_on_{dataset_tag}"
    pair_root.mkdir(parents=True, exist_ok=True)

    print(f"[eval] {model_tag} × {dataset_tag} × K={num_seeds}  →  {pair_root}")

    corpora: list[dict] = []
    per_ex: dict[str, dict] = {}   # id -> {"prompt": ..., "nll": [...], "bpd": [...], ...}

    for k in range(num_seeds):
        save_dir = pair_root / f"seed_{k:02d}"
        print(f"  → seed {k+1}/{num_seeds}")
        run_evaluation(MDLMEvalConfig(
            model_name_or_path=ckpt, gold_path=gold, save_summary_path=str(save_dir),
            eval_seed=k, **FINAL_EVAL_CONFIG,
        ))

        corpora.append(json.loads(next(save_dir.glob("*_corpus.json")).read_text()))
        with open(next(save_dir.glob("*_per_example.jsonl"))) as f:
            for row in map(json.loads, f):
                slot = per_ex.setdefault(row["id"], {"prompt": row.get("prompt", ""),
                                                    **{m: [] for m in _METRIC_KEYS}})
                for m in _METRIC_KEYS:
                    if row.get(m) is not None:
                        slot[m].append(row[m])

    # ── corpus-level summary ─────────────────────────────────────────────
    summary = {
        "model_tag": model_tag, "dataset_tag": dataset_tag,
        "checkpoint": ckpt, "gold_path": gold,
        "num_seeds": num_seeds, "n_examples": corpora[0]["n_examples"],
        "metrics": {m: _mean_se([c[m] for c in corpora]) for m in _METRIC_KEYS},
    }
    (pair_root / "summary.json").write_text(json.dumps(summary, indent=2) + "\n")

    # ── per-example summary ──────────────────────────────────────────────
    with open(pair_root / "per_example_summary.jsonl", "w") as f:
        for ex_id, slot in per_ex.items():
            f.write(json.dumps({
                "id": ex_id, "prompt": slot["prompt"],
                "metrics": {m: _mean_se(slot[m]) for m in _METRIC_KEYS},
            }) + "\n")

    for m in _METRIC_KEYS:
        s = summary["metrics"][m]
        print(f"    {m:22s} = {s['mean']:.4f} ± {s['se']:.4f}  (K={s['k']})")

    return summary

#evaluate_checkpoint("M1", "test")
for m in MODEL_CHECKPOINTS:
    evaluate_checkpoint(m, "test")
```

```python
import json
from pathlib import Path
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import bootstrap

SAVE_ROOT = Path("/content/eval_outputs")

MODEL_CHECKPOINTS = {
    "M0": "/content/mdlm_sft/artifacts_weights_mdlm_base_mdlm-owt_chat",
    "M1": "avgJo3/mdlm-sft-M1-50pct-2ep",
    "M2": "avgJo3/mdlm-sft-M2-100pct-1ep",
}
MODELS  = list(MODEL_CHECKPOINTS)     # ["M0", "M1", "M2"]
ALIAS   = {"M0": "M0", "M1": "M+", "M2": "M++"}
MPP, MP = "M2", "M1"                  # paired-comparison convention: Δ = M++ − M+

N_BOOT, SEED = 10_000, 0

def _load(m):
    rows = []
    with open(SAVE_ROOT / f"{m}_on_test" / "per_example_summary.jsonl") as f:
        for r in map(json.loads, f):
            rows.append({"id": r["id"],
                         f"nll_{m}": r["metrics"]["nll"]["mean"],
                         f"acc_{m}": r["metrics"]["mean_token_accuracy"]["mean"],
                         f"K_{m}":   r["metrics"]["nll"]["k"]})
    return pd.DataFrame(rows)

wide   = _load(MODELS[0])
for m in MODELS[1:]:
    wide = wide.merge(_load(m), on="id")
paired = wide.dropna(subset=[f"{s}_{m}" for s in ("nll", "acc") for m in MODELS]).reset_index(drop=True)
n, K   = len(paired), int(paired[[f"K_{m}" for m in MODELS]].max().max())

def boot_ci(x, fn=np.mean):
    res = bootstrap((x,), fn, n_resamples=N_BOOT, confidence_level=0.95,
                    method="percentile", random_state=SEED, vectorized=False)
    return float(fn(x)), float(res.confidence_interval.low), float(res.confidence_interval.high)

print(f"n = {n}  |  K = {K}  |  dropped = {len(wide) - n}")
```

```python
rows = []
for m in MODELS:
    for label, col in [("NLL", f"nll_{m}"), ("Acc", f"acc_{m}")]:
        mean, lo, hi = boot_ci(paired[col].to_numpy())
        rows.append({"model": ALIAS[m], "metric": label,
                     "cell": f"{mean:.3f} [{lo:.3f}, {hi:.3f}]"})

t1 = pd.DataFrame(rows).pivot(index="model", columns="metric", values="cell") \
       .reindex([ALIAS[m] for m in MODELS])[["NLL", "Acc"]]

print(f"T1 — Marginals (n={n}, K={K}, B={N_BOOT})")
print(t1.to_string())
t1.to_latex(SAVE_ROOT / "t1_marginals.tex", caption=f"Stage-1 marginals ($n={n}$, $K={K}$).", label="tab:stage1_t1")
```

```python
d_nll = (paired[f"nll_{MPP}"] - paired[f"nll_{MP}"]).to_numpy()
d_acc = (paired[f"acc_{MPP}"] - paired[f"acc_{MP}"]).to_numpy()

rows = []
for label, d in [("ΔNLL", d_nll), ("ΔAcc", d_acc)]:
    mean, lo, hi = boot_ci(d)
    _, p         = stats.wilcoxon(d, zero_method="zsplit", alternative="two-sided")
    sk           = float(stats.skew(d, bias=False))
    rows.append({"metric":          label,
                 "mean Δ [95% CI]": f"{mean:+.4f} [{lo:+.4f}, {hi:+.4f}]",
                 "wilcoxon p":      f"{p:.2e}" if p < 1e-3 else f"{p:.4f}",
                 "skew":            f"{sk:+.2f}"})

print(f"T2 — Paired differences (Δ = {ALIAS[MPP]} − {ALIAS[MP]}, n={n}, K={K}, B={N_BOOT})")
print(pd.DataFrame(rows).to_string(index=False))

pd.DataFrame(rows).to_latex(SAVE_ROOT / "t2_paired.tex", index=False,
    caption=f"Stage-1 paired differences $\\Delta = {ALIAS[MPP]} - {ALIAS[MP]}$ ($n={n}$, $K={K}$).",
    label="tab:stage1_t2")
```

```python
rows = []
for label, d, mpp_sign in [("ΔNLL", d_nll, -1), ("ΔAcc", d_acc, +1)]:
    n_mpp = int((np.sign(d) == mpp_sign).sum())
    n_mp  = int((np.sign(d) == -mpp_sign).sum())
    n_tie = int((d == 0).sum())
    rho   = (n_mpp - n_mp) / (n_mpp + n_mp)
    lo_p  = stats.beta.ppf(0.025, n_mpp,     n_mp + 1)
    hi_p  = stats.beta.ppf(0.975, n_mpp + 1, n_mp)
    r_lo, r_hi      = 2 * lo_p - 1, 2 * hi_p - 1
    med, m_lo, m_hi = boot_ci(d, fn=np.median)
    rows.append({"metric":                            label,
                 f"{ALIAS[MPP]}/{ALIAS[MP]}/tie":    f"{n_mpp}/{n_mp}/{n_tie}",
                 "ρ [95% CI]":                       f"{rho:+.4f} [{r_lo:+.4f}, {r_hi:+.4f}]",
                 "median Δ [95% CI]":                f"{med:+.5f} [{m_lo:+.5f}, {m_hi:+.5f}]"})

print(f"T3 — Sign bins, ρ, median (Δ = {ALIAS[MPP]} − {ALIAS[MP]}, n={n})")
print(pd.DataFrame(rows).to_string(index=False))

pd.DataFrame(rows).to_latex(SAVE_ROOT / "t3_signbins.tex", index=False,
    caption=f"Stage-1 sign bins ($n={n}$).", label="tab:stage1_t3")
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

# ── W&B-inspired style: light grid, sans-serif, muted qualitative palette ─────
rcParams.update({
    "font.family":       "sans-serif",
    "font.sans-serif":   ["DejaVu Sans", "Arial", "Helvetica"],
    "font.size":          10,
    "axes.titlesize":     10,
    "axes.labelsize":     9,
    "axes.spines.top":    False,
    "axes.spines.right":  False,
    "axes.edgecolor":     "#4A4A4A",
    "axes.linewidth":     0.8,
    "axes.grid":          True,
    "grid.color":         "#E5E5E5",
    "grid.linewidth":     0.6,
    "xtick.color":        "#4A4A4A",
    "ytick.color":        "#4A4A4A",
    "xtick.labelsize":    8,
    "ytick.labelsize":    8,
    "legend.fontsize":    8,
    "legend.frameon":     False,
    "figure.facecolor":   "white",
    "axes.facecolor":     "white",
    "savefig.facecolor":  "white",
    "savefig.bbox":       "tight",
    "pdf.fonttype":       42,     # embed TrueType (editable in Illustrator, prints crisp)
    "ps.fonttype":        42,
})

# W&B qualitative palette (their default categorical set, colorblind-safe, print-friendly)
COLOR = {
    "M+":     "#2E8BC0",   # W&B blue
    "M++":    "#F2994A",   # W&B orange
    "points": "#5B8FF9",   # softer blue for dense scatter/hist
    "mean":   "#E15759",   # coral red
    "median": "#F2C94C",   # amber
    "ref":    "#2C2C2C",   # near-black axis lines
    "guide":  "#BDBDBD",   # light grey dashed guides
}
POINT_KW = dict(s=6, alpha=0.28, color=COLOR["points"], edgecolor="none", rasterized=True)
LINE_KW  = dict(lw=1.4)
REF_KW   = dict(color=COLOR["ref"],   lw=0.7)
GUIDE_KW = dict(color=COLOR["guide"], lw=0.9, ls=(0, (4, 3)))

# ── shared sizes (thesis full text width ≈ 6.3 in; use 6.3 wide for print) ────
FIGSIZE_WIDE   = (6.3, 2.6)    # 2-panel
FIGSIZE_SQUARE = (6.3, 4.4)    # 1-panel

FIG_DIR = SAVE_ROOT / "figures"; FIG_DIR.mkdir(exist_ok=True)


def _save(fig, name):
    fig.savefig(FIG_DIR / f"{name}.pdf")
    fig.savefig(FIG_DIR / f"{name}.png", dpi=200)


def fig_paired_diff_hist(d_nll, d_acc, *, bins=80):
    """Fig. 1 — Paired-difference distributions (symmetry check)."""
    fig, axes = plt.subplots(1, 2, figsize=FIGSIZE_WIDE)
    for ax, (label, d) in zip(axes, [("ΔNLL", d_nll), ("ΔAcc", d_acc)]):
        ax.hist(d, bins=bins, color=COLOR["points"], alpha=0.85, edgecolor="none")
        ax.axvline(0,            **REF_KW)
        ax.axvline(d.mean(),     color=COLOR["mean"],   **LINE_KW,
                   label=f"mean = {d.mean():+.4f}")
        ax.axvline(np.median(d), color=COLOR["median"], **LINE_KW,
                   label=f"median = {np.median(d):+.5f}")
        ax.set_title(f"{label}   (skew = {stats.skew(d, bias=False):+.2f})")
        ax.set_xlabel(label); ax.set_ylabel("items")
        ax.legend(loc="upper right")
    fig.suptitle(f"Paired-difference distributions  (Δ = {ALIAS[MPP]} − {ALIAS[MP]}, n={n})",
                 fontsize=10, y=1.02)
    fig.tight_layout(); _save(fig, "fig1_paired_diff_hist"); return fig


def fig_bland_altman(paired):
    """Fig. 2 — Bland–Altman: Δⱼ vs mean of the two scores per item."""
    fig, axes = plt.subplots(1, 2, figsize=FIGSIZE_WIDE)
    for ax, (label, key) in zip(axes, [("NLL", "nll"), ("Acc", "acc")]):
        a, b   = paired[f"{key}_{MP}"].to_numpy(), paired[f"{key}_{MPP}"].to_numpy()
        mid, d = 0.5 * (a + b), b - a
        lo, hi = d.mean() - 1.96 * d.std(), d.mean() + 1.96 * d.std()
        ax.scatter(mid, d, **POINT_KW)
        ax.axhline(0,        **REF_KW)
        ax.axhline(d.mean(), color=COLOR["mean"], **LINE_KW,
                   label=f"mean Δ = {d.mean():+.4f}")
        ax.axhline(lo, **GUIDE_KW, label="±1.96 sd")
        ax.axhline(hi, **GUIDE_KW)
        ax.set_xlabel(f"½·({ALIAS[MP]} + {ALIAS[MPP]})  [{label}]")
        ax.set_ylabel(f"Δ{label}")
        ax.set_title(f"Δ{label} vs item difficulty")
        ax.legend(loc="upper right")
    fig.suptitle(f"Bland–Altman: exposure gap vs item difficulty  (n={n})",
                 fontsize=10, y=1.02)
    fig.tight_layout(); _save(fig, "fig2_bland_altman"); return fig


def fig_metric_agreement(d_nll, d_acc):
    """Fig. 3 — Δ_NLL vs Δ_Acc per item; quadrant counts."""
    fig, ax = plt.subplots(figsize=FIGSIZE_SQUARE)
    ax.scatter(d_nll, d_acc, **POINT_KW)
    ax.axhline(0, **REF_KW); ax.axvline(0, **REF_KW)

    q = {"both M++":         ((d_nll < 0) & (d_acc > 0)).sum(),
         "both M+":          ((d_nll > 0) & (d_acc < 0)).sum(),
         "NLL→M++, Acc→M+":  ((d_nll < 0) & (d_acc < 0)).sum(),
         "NLL→M+, Acc→M++":  ((d_nll > 0) & (d_acc > 0)).sum()}
    x_lo, x_hi = ax.get_xlim(); y_lo, y_hi = ax.get_ylim()
    corners = {"both M++":         (x_lo * 0.95, y_hi * 0.92, "left",  "top"),
               "both M+":          (x_hi * 0.95, y_lo * 0.92, "right", "bottom"),
               "NLL→M++, Acc→M+":  (x_lo * 0.95, y_lo * 0.92, "left",  "bottom"),
               "NLL→M+, Acc→M++":  (x_hi * 0.95, y_hi * 0.92, "right", "top")}
    for k, (x, y, ha, va) in corners.items():
        ax.text(x, y, f"{k}\nn = {q[k]}", ha=ha, va=va, fontsize=8,
                bbox=dict(boxstyle="round,pad=0.3", fc="white",
                          ec=COLOR["guide"], lw=0.8, alpha=0.95))

    ax.set_xlabel(f"ΔNLL   ← {ALIAS[MPP]} better | {ALIAS[MP]} better →")
    ax.set_ylabel(f"ΔAcc   ← {ALIAS[MP]} better | {ALIAS[MPP]} better →")
    ax.set_title(f"Per-item agreement of NLL and accuracy  (n={n})")
    fig.tight_layout(); _save(fig, "fig3_metric_agreement"); return fig


fig_paired_diff_hist(d_nll, d_acc)
fig_bland_altman(paired)
fig_metric_agreement(d_nll, d_acc)
plt.show()
print(f"figures saved to {FIG_DIR}")
```

```python
SWEEP_CONFIG = {
    "method": "bayes",
    "metric": {"goal": "minimize", "name": "eval/nll"},
    "parameters": {
        # ── SWEPT ──
        "learning_rate": {
            "distribution": "log_uniform_values",
            "min": 5.0e-4,   # ← updated from 3e-4
            "max": 5.0e-3,
        },

        # ── PINNED at Stage-A top-15 empirical winners ──
        "warmup_ratio":   {"value": 0.1},    # ← updated from 0.03 (60% of top-15)
        "weight_decay":   {"value": 0.01},   # ← rank-1; top-15 uniform
        "time_epsilon":   {"value": 1e-3},

        # ── Substrate: 50% ──
        "max_steps":                   {"value": 2344},
        "eval_steps":                  {"value": 469},

        # ── Training procedure pins ──
        "lr_scheduler_type":           {"value": "cosine"},
        "per_device_train_batch_size": {"value": 64},
        "per_device_eval_batch_size":  {"value": 64},
        "gradient_accumulation_steps": {"value": 5},
        "save_strategy":               {"value": "no"},
        "loss_weight_type":            {"value": "scheduler"},
    },
}

FIXED = {
    "load_model_train_path": "/content/mdlm_sft/artifacts_weights_mdlm_base_mdlm-owt_chat",
    "load_data_train_path":  "/content/cot-strat/strat_train_50pct",
    "load_data_eval_path":   "/content/cot-strat/validation",
}

# ── Run-name schema ───────────────────────────────────────────────────────
SWEPT = {
    "learning_rate": "{:.0e}",
    "warmup_ratio":  "{:.2f}",
    "weight_decay":  "{:.2f}",       # → wd0.00, wd0.01, wd0.10
}

def build_run_name(sweep_kwargs: dict) -> str:
    return "-".join(["sweep"] + [
        f"{''.join(w[0] for w in k.split('_'))}{fmt.format(sweep_kwargs[k])}"
        for k, fmt in SWEPT.items() if k in sweep_kwargs
    ])

def sweep_entry():
    with wandb.init() as run:
        sweep_kwargs = dict(wandb.config)
        round_name = build_run_name(sweep_kwargs)
        run.name = round_name
        save_model_train_path = f"/model_out_{round_name}"

        print(f"[{round_name}] run={run.id}")
        print(f"  hyperparameters: {sweep_kwargs}")

        try:
            cfg = MDLMSFTConfig(
                model_name_or_path=FIXED["load_model_train_path"],
                output_dir=save_model_train_path,
                train_ds_path=FIXED["load_data_train_path"],
                eval_ds_path=FIXED["load_data_eval_path"],
                run_name=f"mdlm-sft-{round_name}",
                **sweep_kwargs,
            )
            run_training(cfg)
        except Exception as e:
            print(f"[{round_name}] FAILED: {e!r}")
            raise

def main(count: int = 1, project: str = "huggingface"):
    pprint.pprint(SWEEP_CONFIG)
    sweep_id = wandb.sweep(SWEEP_CONFIG, project=project)
    print(f"sweep_id = {sweep_id}")
    wandb.agent(sweep_id, sweep_entry, count=count)

if __name__ == "__main__":
    main(count=24)
```

```python
"""Plot Stage 1 training dynamics for M1 and M2 from W&B."""

from pathlib import Path
import matplotlib.pyplot as plt
import wandb

ENTITY, PROJECT = "AvgJ02", "huggingface"
RUNS = {"M1 (repetition)": "mdlm-sft-final-M1", "M2 (breadth)": "mdlm-sft-final-M2"}
COLOURS = {"M1 (repetition)": "#FF9838", "M2 (breadth)": "#3B7DDD"}
OUT = Path(".")

PANELS = [
    ("train/loss",                "train/loss"),
    ("eval/loss",                 "eval/loss"),
    ("train/entropy",             "train/entropy"),
    ("eval/entropy",              "eval/entropy"),
    ("train/mean_token_accuracy", "train/mean_token_accuracy"),
    ("eval/mean_token_accuracy",  "eval/mean_token_accuracy"),
    ("eval/nll",                  "eval/nll"),
    ("eval/bpd",                  "eval/bpd"),
]

plt.rcParams.update({
    "figure.facecolor": "white", "axes.facecolor": "#f5f5f5",
    "axes.edgecolor": "#cccccc", "axes.grid": True,
    "grid.color": "white", "grid.linewidth": 0.8,
    "axes.spines.top": False, "axes.spines.right": False,
    "axes.labelsize": 9, "axes.titlesize": 10,
    "xtick.labelsize": 8, "ytick.labelsize": 8,
    "xtick.color": "#666666", "ytick.color": "#666666",
    "font.family": "sans-serif", "font.sans-serif": ["DejaVu Sans"],
    "lines.linewidth": 1.3, "legend.frameon": False, "legend.fontsize": 8,
})

api = wandb.Api()
dfs = {}
for label, name in RUNS.items():
    run = api.runs(f"{ENTITY}/{PROJECT}", filters={"display_name": name})[0]
    print(f"[fetch] {label} -> {run.id}")
    df = run.history(samples=100_000, pandas=True)
    df.to_csv(OUT / f"{name}.csv", index=False)
    dfs[label] = df

def plot_panel(ax, title, key, legend=False):
    for label, df in dfs.items():
        sub = df[["_step", key]].dropna()
        ax.plot(sub["_step"], sub[key], color=COLOURS[label], label=label)
    ax.set_title(title); ax.set_xlabel("step")
    if legend: ax.legend(loc="best")

fig, axes = plt.subplots(4, 2, figsize=(8.5, 10), sharex=True)
for ax, (title, key) in zip(axes.flat, PANELS):
    plot_panel(ax, title, key, legend=(ax is axes[0, 0]))
fig.suptitle("Stage 1 training dynamics: M+ vs. M++", y=0.995)
fig.tight_layout(rect=(0, 0, 1, 0.985))
fig.savefig(OUT / "training_dynamics_grid.pdf", bbox_inches="tight")
plt.close(fig)
print("[write] training_dynamics_grid.pdf")

for title, key in PANELS:
    fig, ax = plt.subplots(figsize=(4.2, 3.0))
    plot_panel(ax, title, key, legend=True)
    fig.tight_layout()
    out = OUT / f"tile_{title.replace('/', '_')}.pdf"
    fig.savefig(out, bbox_inches="tight")
    plt.close(fig)
    print(f"[write] {out.name}")

print("done.")
```
